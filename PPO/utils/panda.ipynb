```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
batch_total_actors_reward, total_loss, zero,uno,due,tre = [], [], [], [], [], [], 
for i in range(len(df)):
    if df.values[i][0] == 'batch_total_actors_reward':
        batch_total_actors_reward.append(df.values[i][1])
    elif  df.values[i][0] == 'total_loss':
        total_loss.append(df.values[i][1])
    elif  df.values[i][0] == "'0'":
        zero.append(df.values[i][1])
    elif  df.values[i][0] == "'1'":
        uno.append(df.values[i][1])
    elif  df.values[i][0] == "'2'":
        due.append(df.values[i][1])
    elif  df.values[i][0] == "'3'":
        tre.append(df.values[i][1])
```

```python
plt.plot(batch_total_actors_reward, '-', color='blue', label='Total actors reward in batch')

plt.title('20s/1k batch - total batch reward', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("1.png")
plt.show()
```

```python
plt.plot(zero, '-', color='lightgreen', label='Actor 0 rewards')
plt.plot(uno, '-',  color='lightblue', label='Actor 1 rewards')
plt.plot(due, '-',  color='orange', label='Actor 2 rewards')
plt.plot(tre, '-',  color='pink', label='Actor 3 rewards')
plt.title('20s/1k batch - agents reward', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("2.png")
plt.show()
```
