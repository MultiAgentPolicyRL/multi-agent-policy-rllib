```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
ARG = "2022-12-20 21:30:37.740486"

actors = pd.read_csv(f'../logs/{ARG}_actors_rew.csv')
loss = pd.read_csv(f'../logs/{ARG}_loss.csv')
total = pd.read_csv(f'../logs/{ARG}_total_rew.csv')
```

```python
# data_logger.info("agent0, agent1, agent2, agent3, total_loss, policy_loss, vf_loss, entropy, total_reward")
# keys: totalreward

plt.plot(total.totalreward, '-', color='blue', label='Total actors reward in batch')

plt.title('20s/1k batch - total batch reward', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("total_batch_reward.png")
plt.show()
```

```python
plt.plot(actors.agent0, '-', color='lightgreen', label='Actor 0 rewards')
plt.plot(actors.agent1, '-',  color='lightblue', label='Actor 1 rewards')
plt.plot(actors.agent2, '-',  color='orange', label='Actor 2 rewards')
plt.plot(actors.agent3, '-',  color='pink', label='Actor 3 rewards')
plt.title('20s/1k batch - agents reward', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("agents_reward.png")
plt.show()
```

```python
totalloss=[]

i=0
while i < len(loss.totalloss):
    totalloss.append((loss.totalloss[i]+loss.totalloss[i+1]+loss.totalloss[i+2]+loss.totalloss[i+3])/4)

    i+=4

plt.plot(totalloss, '-', color='blue', label='Total loss')
plt.title('20s/1k batch - Total Loss', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("total_loss.png")
plt.show()
```

```python
policyloss=[]
vfloss = []
entropy = []
i=0
while i < len(loss.policyloss):
    policyloss.append((loss.policyloss[i]+loss.policyloss[i+1]+loss.policyloss[i+2]+loss.policyloss[i+3])/4)
    vfloss.append((loss.vfloss[i]+loss.vfloss[i+1]+loss.vfloss[i+2]+loss.vfloss[i+3])/4)
    entropy.append((loss.entropy[i]+loss.entropy[i+1]+loss.entropy[i+2]+loss.entropy[i+3])/4)
    
    
    i+=4

# policyloss, vfloss, entropy
plt.plot(policyloss, '-', color='lightblue', label='Policy loss')
plt.plot(vfloss, '-', color='cyan', label='VF loss')
plt.plot(entropy, '--', color='lightcoral', label='Entropy')
plt.title('20s/1k batch - VF Loss, Policy Loss and entropy', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("multiple_losses.png")
plt.show()
```

```python
plt.plot(policyloss, '-', color='lightblue', label='Policy loss')
plt.plot(entropy, '--', color='lightcoral', label='Entropy')
plt.title('20s/1k batch - Entropy and Policy Loss', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("policy_loss_entropy.png")
plt.show()
```

```python
plt.plot(policyloss, '-', color='lightblue', label='Policy loss')
plt.title('20s/1k batch - Policy Loss', fontdict=None, loc='center', pad=None)
plt.legend()
plt.savefig("policy_loss.png")
plt.show()
```
