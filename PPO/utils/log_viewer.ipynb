```python
f=open('experiment_1669828753.160924.txt')

actor_loss, critic_loss = [],[]

for line in f:
    splitted_line = line.split()
    if len(splitted_line)>3:
        # print(splitted_line)
        # print(splitted_line[3])
        if (splitted_line[3] == 'loss:'):
            if (splitted_line[2] == 'Actor'):
                actor_loss.append(float(splitted_line[4]))
            elif (splitted_line[2] == 'Critic'):
                critic_loss.append(float(splitted_line[4]))

y_map = [i for i in range(len(actor_loss))]
```

```python
import matplotlib.pyplot as plt

plt.plot(y_map, actor_loss, label = "Actor Loss")
# plt.plot(y_map, critic_loss, label = "Critic Loss")

plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Actor loss in 20 steps, 250 batch_size")
plt.legend()
# plt.show()
plt.savefig("actor_loss1.png")
```

```python
plt.plot(y_map, critic_loss, label = "Critic Loss")

plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Critic loss in 20 steps, 250 batch_size")
plt.legend()
# plt.show()
plt.savefig("critic_loss1.png")
```
