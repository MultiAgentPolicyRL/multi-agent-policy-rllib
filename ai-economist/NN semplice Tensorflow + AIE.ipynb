```python
def dict_to_array(data):
    array = []
    if type(data) == dict:
        for da in data:
            if type(data[da]) == dict:
                dict_to_array(data[da])
                
            elif type(data[da]) == list:
                for element in data[da]:
                    array.append(element)
            else:
                array.append(data[da])
    elif type(data) == list:           
        for element in data[da]:
            array.append(element)        
    else:
        array.append(data)   
    return array
```
